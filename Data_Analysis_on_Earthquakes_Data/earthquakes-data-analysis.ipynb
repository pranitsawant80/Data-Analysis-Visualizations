## Code to create new column called  and location of earthquake distance from location

```python
def seplocation(n):
    return n.split('-')[1].strip()
```

```python
df['title']=df['title'].apply(seplocation)
```

```python
df.rename(columns={'title':'locearthquake'},inplace=True)
df.head(3)
```

```python
def distlocation(n):
    n.strip()
    d=n.split(' ')[0]
    if ('k' in d)==True:
        d1=d.split('k')[0]
        return d1
    else:
        return d
distlocation('18km SW of Malango, Solomon Islands')
```

```python
df['dist_from_location']=df['locearthquake'].apply(distlocation)
df.head(10)
```

```python
print(list(df.dist_from_location.unique()))
```

```python
def intmake(n):    
    try:
        if type(int(n))==int:
            d=int(n)
            return d
    except:
        d1=None
        return d1
```

```python
df['dist_from_location']=df['dist_from_location'].apply(intmake)
```

## To check correlation between variables

### Converting categorical column to numerical

```python
# To convert dataframe to all numerical values in order to see correlation between those.
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df1=df.drop(columns=['locearthquake','country','location','date_time'],axis=1)
df_cat=df1[['magType','net','alert']]
df_num=df1[['magnitude','cdi','mmi','tsunami','sig','nst','dmin','gap','depth','latitude','longitude','dist_from_location']]
```

```python
for i in df_cat:
    df_cat[i]=le.fit_transform(df_cat[i])
df_cat.head(2)
```

```python
df_new=pd.concat([df_cat,df_num],axis=1)
df_new.head(3)
```

## Plot correlation matrix

```python
plt.figure(figsize=(15, 10))
sns.heatmap(df_new.corr(), annot=True, cmap='coolwarm')
# Display the plot
plt.title('Correlation Matrix')
plt.show()
```
